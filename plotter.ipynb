```python
import numpy as np
from functions.misc.confidence_bounds import bootstrap_ci
import matplotlib.pyplot as plt
from functions.misc.plot_data import plot_data


names = ['fourier', 'legendre', 'poly']#['lasso01', 'lasso001', 'lasso0001']
new_names = ['fourier', 'legendre', 'poly']#['ridge01', 'ridge001', 'ridge0001']
dir = 'results\_24_06_11-18_45_'+'/'

for i in range(len(names)):
    name = names[i]
    results = np.load(dir+name+'.npy')
    low, high = bootstrap_ci(results)

    plot_data(np.arange(0,len(low)), low, high, col='C{}'.format(i+1), label=new_names[i])

plt.legend()
plt.title('Reward curves')
plt.savefig('reward_plot.pdf')
```
